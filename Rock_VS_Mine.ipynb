```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Copy of sonar data.csv', header=None)
data.head()
```

```python
data.rename({60:'Target'}, axis=1, inplace=True)
```

```python
data.head(1)
```

```python
X = data.drop(['Target'], axis=1)
y = data['Target']
```

```python
print("data X",X.shape)
print("data y",y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
train_data_accuracy = accuracy_score(y_train, X_train_prediction) 
```

```python
print("Acuraccy model pada prediksi data x test = ", train_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(y_test,X_test_prediction) 
```

```python
print("Acuraccy model pada prediksi data x test = ", test_data_accuracy)
```
